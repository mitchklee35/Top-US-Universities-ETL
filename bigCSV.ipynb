```python
# Save file path to variable
bigColData_csv = "resources/college_data.csv"
import pandas as pd
```

```python
# Read with Pandas
bigColData_raw = pd.read_csv(bigColData_csv, low_memory=False)
pd.options.display.max_columns = 2000
bigColData_raw
```

```python
INSTNM = bigColData_raw ['INSTNM']
CITY = bigColData_raw ['CITY']
STABBR = bigColData_raw ['STABBR']
INSTURL = bigColData_raw ['INSTURL']
OMENRUP_PARTTIME_POOLED_SUPP = bigColData_raw ['OMENRUP_PARTTIME_POOLED_SUPP']
FTFTPCTPELL = bigColData_raw ['FTFTPCTPELL']
FTFTPCTFLOAN = bigColData_raw ['FTFTPCTFLOAN']
UG12MN = bigColData_raw ['UG12MN']
G12MN = bigColData_raw ['G12MN']
SCUGFFN = bigColData_raw ['SCUGFFN']
POOLYRS_FTFTAIDPCT = bigColData_raw ['POOLYRS_FTFTAIDPCT']
FTFTPCTPELL_POOLED_SUPP = bigColData_raw ['FTFTPCTPELL_POOLED_SUPP']
FTFTPCTFLOAN_POOLED_SUPP = bigColData_raw ['FTFTPCTFLOAN_POOLED_SUPP']
SCUGFFN_POOLED = bigColData_raw ['SCUGFFN_POOLED']
ADM_RATE_ALL = bigColData_raw ['ADM_RATE_ALL']
SAT_AVG_ALL = bigColData_raw ['SAT_AVG_ALL']
COSTT4_A = bigColData_raw ['COSTT4_A']
TUITIONFEE_IN = bigColData_raw ['TUITIONFEE_IN']
TUITIONFEE_OUT = bigColData_raw ['TUITIONFEE_OUT']
UGDS = bigColData_raw ['UGDS']
UGDS_WHITE = bigColData_raw ['UGDS_WHITE']
UGDS_BLACK = bigColData_raw ['UGDS_BLACK']
UGDS_HISP = bigColData_raw ['UGDS_HISP']
UGDS_ASIAN = bigColData_raw ['UGDS_ASIAN']
UGDS_AIAN = bigColData_raw ['UGDS_AIAN']
UGDS_NHPI = bigColData_raw ['UGDS_NHPI']
UGDS_2MOR = bigColData_raw ['UGDS_2MOR']
UGDS_NRA = bigColData_raw ['UGDS_NRA']
UGDS_UNKN = bigColData_raw ['UGDS_UNKN']
```

```python
bigColData_df = pd.DataFrame({'Institution Name':INSTNM,
                             'City':CITY,
                             'State':STABBR,
                             'URL':INSTURL,
                              'In-state tuition & fees':TUITIONFEE_IN,
                              'Out-of-state tuition & fees':TUITIONFEE_OUT,
                              'Avg cost of attendance/4ys':COSTT4_A,
                              '% FT Pell':FTFTPCTPELL,
                              '% FT Stu Loan':FTFTPCTFLOAN,
                              'Admission Rate':ADM_RATE_ALL,
                              'Avg SAT score ':SAT_AVG_ALL,
                              'Total undergrad enrollment':UGDS,
                              'Undergrad ethnicity-white':UGDS_WHITE,
                              'Undergrad ethnicity-black':UGDS_BLACK,
                              'Undergrad ethnicity-hispanic':UGDS_HISP,
                              'Undergrad ethnicity-asian':UGDS_ASIAN,
                              'Undergrad ethnicity-American Indian':UGDS_AIAN,
                              'Undergrad ethnicity-Native Hawaiian/Pacific Islander':UGDS_NHPI,
                              'Undergrad ethnicity-':UGDS_2MOR,
                              'Undergrad ethnicity-':UGDS_NRA,
                              'UGDS_NRA':UGDS_NRA,
                              'UGDS_UNKN':UGDS_UNKN,
                              'Undergrad enrollment, last 12 mo':UG12MN,
                              'Grad enrollment, last 12 mo':G12MN,
})

bigColData_df
```

```python
big_coll_df = bigColData_df.dropna()
college_condensed = big_coll_df[['Institution Name','City','State','Out-of-state tuition & fees','Admission Rate','Avg SAT score ','Total undergrad enrollment']]
college_df = college_condensed.rename(columns={"Institution Name":"Name"})
college_df
```

```python
rankings_df = pd.read_csv("resources/college_rankings_kaggle.csv")
US_ranking = rankings_df.loc[rankings_df.country == "USA"]
US_ranking_2015 = US_ranking.loc[US_ranking.year == 2015]
US_ranking_2015.dropna()
US_df = US_ranking_2015.rename(columns={'institution':'Name'})
```

```python
merged_df = pd.merge(college_df, US_df,on="Name",how="right" )
merged_2 = merged_df.fillna("None")
merged_2.loc[merged_2.City == "None"]
```
